```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread("/content/abstract-colourful-rectangular-texture-background-vector-30709699.jpg")
img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
img=cv2.resize(img,(250,350))
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
X = img.reshape((-1,3)) #to reshape 2D image into array or a 1D
print(X.shape)
```

```python
X
```

First

```python
from sklearn.cluster import KMeans
```

```python
k=3
model=KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids=model.cluster_centers_
```

```python
centroids
```

```python
import numpy as np
colours=np.array(centroids,dtype='uint8')
```

```python
mat = np.zeros((100,100,3),dtype='uint8')
plt.imshow(mat)
plt.show()
```

```python
mat[:,:,:]=100
```

```python
i = 1
for color in colours:
  plt.subplot(1,k,i)
  plt.axis("off")
  i = i+ 1
  mat = np.zeros((100,100,3),dtype='uint8')
  mat[:,:,:] = color
  plt.imshow(mat)
plt.show()
```
